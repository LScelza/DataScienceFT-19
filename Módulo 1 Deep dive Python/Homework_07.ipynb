Voy a hacer un arbol tipo Max-Heap, en el que siempre los valores menores se van ordenando hacia la izquierda.

```python
import pandas as pd
import random
import numpy as np

```

Empiezo creando la clase Nodo. Cada nodo, a mi entender, tiene que tener un valor, un nivel y tiene que saber los valores que tiene a la izquierda y a la derecha (hijos) y el valor del nivel superior (padre) y tiene que poder devolver estos valores. Varios atributos los voy a dejar públicos para que se puedan modificar desde la clase Arbol. Los métodos los voy a incluir en Arbol.

```python
class Nodo():
    
    def __init__(self, valor:int) -> None:
        self.__valor = valor
        self.nivel = None
        self.izquierda = None
        self.derecha = None
        self.padre = None

    @property
    def valor(self) -> int:
        return self.__valor
    
    @property
    def nivel(self) -> int:
        return self.__nivel
    
    @nivel.setter
    def nivel(self, valor:int) -> None:
        self.__nivel = valor

    def es_hijo(self) -> bool:
        if self.padre != None:
            return True
        return False

    def es_padre(self) -> bool:
        if self.derecha != None or self.izquierda != None:
            return True
        return False
    
       
    
    
```

Aclaración: Nodo.es_hijo() y Nodo.es_padre() están solo para después graficar. No hice el gráfico en este ejercicio.

Ahora voy a crear la clase Arbol, que va a contener los métodos de búsqueda, para insertar un nodo, etc.

Importante: La convención para enumerar los nodos es de izquierda a derecha y de arriba a abajo. Se puede comenzar con 0 o con 1. Yo voy a tomar la raíz como 0. Si tomamos 0, la distancia a un hijo izquierdo es 2 * n + 1 y a un hijo derecho 2 * n + 2. La distancia de un hijo al padre es (n -1) // 2.

```python
class Arbol:

    def __init__(self) -> None:
        self.__raiz = None
        self.__puntero = None
        self.__dict_nodos = {'nodo':[], 'nivel':[], 'valor':[], 'padre':[], 'izquierda':[], 'derecha':[]}
        self.__nodos = None
      
    @property
    def nodos(self) -> pd.DataFrame:
        return self.__nodos
    
    def buscaVal(self, valor:int) -> bool:
        if valor in self.__dict_nodos['valor']:
            return True
        return False
        
    def insertaVal(self, valor:int) -> None:
        n = Nodo(valor)
        self.__dict_nodos['nodo'].append(n)
        self.__dict_nodos['nivel'].append(n.nivel)
        self.__dict_nodos['valor'].append(n.valor)
        self.__dict_nodos['padre'].append(n.padre)
        self.__dict_nodos['izquierda'].append(n.izquierda)
        self.__dict_nodos['derecha'].append(n.derecha)
        self.actualizar_df()
        self.ordenar_1_vuelta()
        self.asignar_niveles()
        self.ordenar_2_vuelta()
        self.asignar_enlaces()        
    
    def actualizar_df(self) -> None:
        self.__nodos = pd.DataFrame(self.__dict_nodos)
        self.__nodos.index.name = 'numero'
    
    def ordenar_1_vuelta(self) -> None:
        self.__nodos.sort_values('valor', ascending=False, inplace=True)
        self.__nodos.reset_index(drop=True, inplace=True)
    
    def ordenar_grupo(self, grupo:pd.DataFrame) -> pd.DataFrame:
          return grupo.sort_values(by='valor', ascending=True)

    def ordenar_2_vuelta(self) -> None:
        self.__nodos = self.__nodos.groupby('nivel', group_keys=False, sort=False).apply(self.ordenar_grupo)
        self.__nodos.reset_index(drop=True, inplace=True)
        
    def asignar_niveles(self) -> None:
        potencia = 0
        for i in range(len(self.__nodos)):
            if i == 2 ** potencia - 1:
                potencia += 1
            self.__nodos.loc[i, 'nivel'] = potencia - 1
     
    def asignar_enlaces(self) -> None:
        for i, nodo in self.__nodos.iterrows():
            izq = 2 * i + 1
            der = 2 * i + 2
            try:
                self.__nodos.loc[i, 'izquierda'] = self.__nodos.loc[izq, 'valor']
                self.__nodos.loc[i, 'derecha'] = self.__nodos.loc[der, 'valor']
            except:
                continue  
        for i, nodo in self.__nodos.iterrows():
            if i > 0:
                indice_padre = (i - 1) // 2
                self.__nodos.loc[i, 'padre'] = self.__nodos.loc[indice_padre, 'valor']

    def verVal(self) -> None:
        return self.__nodos 

    


```
